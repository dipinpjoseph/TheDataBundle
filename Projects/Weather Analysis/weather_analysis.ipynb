### Suggest appropriate time for Rain Water Drainage Maintenance Department in Maryland, Texas, Colorado and New York to start Maintenance work.

```python
def month_ext(mon):
    return(mon.split('-')[1])
    
w_data_sel['wh_month'] = w_data_sel['wh_date'].apply(lambda x: month_ext(x))
# Lesser the Dew Point temperature, better work environment
w_data_sel[w_data_sel['wh_dewpoint_temp']<7]['wh_month'].value_counts()
```
