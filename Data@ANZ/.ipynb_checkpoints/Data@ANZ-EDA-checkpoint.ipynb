## Data Pre-processing

```python
# Loading libraries
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading dataset from xlsx file
#!pip install xlrd
df = pd.read_excel("ANZ synthesised transaction dataset.xlsx")
df.head()
```

```python
# Duplicate row check
print(len(df))
print(len(df.drop_duplicates()))
```

```python
# Basic statistics of dataframe
df.describe()
```

```python
df.dtypes
```

```python
# Missing Values
df.isnull().sum()
```

```python
df[df['merchant_id'].isnull()].head()
```

```python
# Transactions types with and without merchants
print(df[df['merchant_id'].isnull()].txn_description.value_counts())
print(df[~df['merchant_id'].isnull()].txn_description.value_counts())
```

```python
# Extracting latitude and longitude of acccount holders
geo = df["long_lat"].str.split("-", n = 1, expand = True)
df["lat"] = pd.to_numeric(geo[1], errors='coerce')
df["lon"] = pd.to_numeric(geo[0], errors='coerce')
#df.drop(columns =["long_lat"],inplace=True)

# Extracting latitude and longitude of acccount holders
geo = df["merchant_long_lat"].str.split("-", n = 1, expand = True) 
df["merch_lat"] = pd.to_numeric(geo[1], errors='coerce')
df["merch_lon"] = pd.to_numeric(geo[0], errors='coerce')
#df.drop(columns =["merchant_long_lat"],inplace=True)

df.head()
```

## Data Visualization

```python
df_map1 = df.groupby(['lon','lat','account'])['balance'].sum()
df_map2 = df.groupby(['merch_lon','merch_lat','account'])['balance'].sum()
```
