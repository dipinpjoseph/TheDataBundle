### Join Merge Concatenation in Pandas

```python
import pandas as pd
df1 = pd.DataFrame({'val': ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'num': range(7)})
df1
```

```python
df2 = pd.DataFrame({'val': ['b', 'a']})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.merge(df1,df2)
```

```python
df1.join(df2.set_index('val'))
```

```python
df1.join(df2.set_index('val'))
```

```python
pd.merge(df1,df2)
```

```python
pd.concat([df1,df2])
```

```python
df1.join(df2,how='left', lsuffix='_left', rsuffix='_right')
```
